```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn import feature_extraction, linear_model, model_selection, preprocessing
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import TweetTokenizer
import string
```

```python
def preprocess(df):
    df = df.copy()

    def clean_text(x):
      # remove new lines
      # remove punctuation
      punctuation = string.punctuation.replace("#", "")
      exclude = set(punctuation)
      x = ''.join(ch for ch in x if ch not in exclude)
      return x

    def tokenize(x):
        tokenizer = TweetTokenizer()
        tokens = tokenizer.tokenize(x.lower())
        return tokens

    df["text"] = df["text"].apply(clean_text)
    df["text"] = df["text"].apply(tokenize)
    return df
```

```python
df_train_loaded = pd.read_csv('/content/drive/MyDrive/2-folder/kaggle/NLP-with-disaster-tweets/train.csv')
df_test_loaded = pd.read_csv('/content/drive/MyDrive/2-folder/kaggle/NLP-with-disaster-tweets/test.csv')
print(df_train_loaded.head(5))

# copy train and test dataframes
df_train = df_train_loaded.copy()
df_test = df_test_loaded.copy()

df_train = preprocess(df_train)
print(df_train.iloc[1]["text"])
```
